* PLA算法是用来二分类的一种算法，y取值只有+1和-1


```python
import numpy as np
import operator

def createTrainDataSet():
    trainDataSet = [[1,  1, 4], 
                    [1,  2, 3], 
                    [1, -2, 3], 
                    [1, -2, 2], 
                    [1,  0, 1], 
                    [1,  1, 2]]
    trainLabel = [1, 1, 1, -1, -1,  -1]
    return trainDataSet, trainLabel

def createTestDataSet():
    testDataSet = [[1, 1, 1], 
                   [1, 2, 0], 
                   [1, 2, 4], 
                   [1, 1, 3]]
    return testDataSet

def pla(dataSet, labels):
    dataMat = np.mat(dataSet)
    labelsMat = np.mat(labels).transpose()
    m,n = np.shape(dataMat)
    weights = np.ones((n,1))
    alpha = 0.1
    while True:
        Done = True
        for i in range(m):
            if (dataMat[i] * weights) * labelsMat[i] > 0:
                continue
            else:
                weights += (alpha * dataMat[i].transpose() * labelsMat[i])
                Done = False
        if Done:
            break
    return weights

dataSet, labels = createTrainDataSet()
weights = pla(dataSet, labels)
print(weights)
```
